```python
from matminer.datasets.dataset_retrieval import load_dataset
import pandas as pd
from MatFold import MatFold

df = load_dataset('matbench_jdft2d')
df.head()   

```

```python
from pymatgen.core import Structure

df['mat_index'] = ['mat{}'.format(i) for i in range(len(df))]
df = df[['mat_index', 'structure', 'exfoliation_en']]
df['composition'] = df.structure.apply(lambda x: x.composition)
mat_struct_dict = {
    row['mat_index']: row['structure'].as_dict() for _, row in df.iterrows()
}
print(df.head())
```

```python
df_clean=df.drop(columns=['structure', 'composition'])
mf = MatFold(df_clean, mat_struct_dict)
#print("composition:", mf.split_statistics("composition"))
#print("chemsys:", mf.split_statistics("chemsys"))
#print("sgnum:", mf.split_statistics("sgnum"))
#print("pointgroup:", mf.split_statistics("pointgroup"))
#print("crystalsys:", mf.split_statistics("crystalsys"))
#print("elements:", mf.split_statistics("elements"))
#print("periodictablerows:", mf.split_statistics("periodictablerows"))
print("periodictablegroups:", mf.split_statistics("periodictablegroups"))

```

```python
train_df, val_df, test_df = mf.create_train_validation_test_splits("random", "elements", 
                                                                   train_fraction=0.7, 
                                                                   validation_fraction=0.2, 
                                                                   test_fraction=0.1,
                                                                   n_test_min=4)


```
